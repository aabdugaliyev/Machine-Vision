```python
import matplotlib.pyplot as plt
import matplotlib.image as pltimg
im = pltimg.imread('.\\6_26.jpg')
fig = plt.figure(figsize=(15,10))
plt.subplot(221)
plt.imshow(im[:,:,0])
plt.subplot(222)
plt.imshow(im[:,:,1])
plt.subplot(223)
plt.imshow(im[:,:,2])
plt.subplot(224)
plt.imshow(im)
plt.show()
```
