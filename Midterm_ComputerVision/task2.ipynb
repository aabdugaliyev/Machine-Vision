```python
import matplotlib.pyplot as plt
import matplotlib.image as pltimg
import numpy as np
import matplotlib.image as img
im = pltimg.imread('tiger.jpg')
fig = plt.figure()
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.114])

img = pltimg.imread('tiger.jpg')     
gray = rgb2gray(img)    

fig = plt.figure()
plt.subplot(221)
plt.plot(im[:,500,1])
plt.subplot(222)
plt.imshow(im)
plt.subplot(223)
plt.imshow(gray, cmap = plt.get_cmap('gray'))
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.misc as img
from scipy import ndimage as nimg


im = img.imread("tiger.jpg")
startIm = im
fig = plt.figure(figsize = (10, 15))
plt.subplot(331)
plt.imshow(startIm)

sumRed = 0
sumGreen = 0
sumBlue = 0
for i in range (im.shape[0]):
    for j in range(im.shape[1]):
        sumRed += im[i, j, 0]
        sumGreen += im[i, j, 1]
        sumBlue += im[i, j, 2]
        
mRed = sumRed / im.size
mGreen = sumGreen / im.size
mBlue = sumBlue / im.size

avg = (mRed + mGreen + mBlue) / 3

for i in range(im.shape[0]):
    for j in range(im.shape[1]):
        im[i, j, 0] = im[i, j, 0] * avg / mRed
        im[i, j, 1] = im[i, j, 1] * avg / mGreen
        im[i, j, 2] = im[i, j, 2] * avg / mBlue

        

plt.subplot(332)
plt.plot(im[:, 500, 1])        
plt.subplot(333)
plt.imshow(im)
plt.show()
```
